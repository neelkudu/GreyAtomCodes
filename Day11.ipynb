```python
df=df.drop(["instant","holiday"],1)
```

```python
df.head()
```

```python
cols=list(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_scaled=pd.DataFrame(scaler.fit_transform(df))
```

```python
df_scaled.columns=cols
```

### Feature Selection using Filter method for Linear Regression

```python
df_scaled.corr()
```

```python
best_features=["casual","registered"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=df_scaled[best_features]
y=df["cnt"]
```

```python
lin_reg = LinearRegression(normalize=True)
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train, y_test = tts(X,y, test_size = 0.3, random_state = 42)
```

```python
lin_reg.fit(X_train,y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

### Backward Elimination method

```python
import statsmodels.api as sm
```

```python
X=df_scaled.drop(['cnt'],1)
y=df["cnt"]
```

```python
X_1=sm.add_constant(X)
```

```python
model_1=sm.OLS(y,X_1).fit()
model_1.summary()
```
